```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("globalterrorismdb_0718dist.csv", encoding='latin1')
```

```python
df.head()
```

```python
# here we are intrested in some major columns so we selected that ones
df1=df[['eventid', 'iyear', 'imonth', 'iday', 'extended',
       'resolution', 'country', 'country_txt', 'region', 'region_txt',
       'provstate', 'city', 'latitude', 'longitude','attacktype1_txt','weaptype1_txt','nkill','gname']]
```

```python
df1.head()
```

```python
df1.hist(figsize=(20,30))
```

```python
sns.heatmap(np.round(df1.corr(),2),annot=True, cmap='BuPu')
```

```python
pd.crosstab(df1.iyear,df1.region_txt).plot(kind='area', figsize=(10,6))
plt.title('Activites by terrorists in region vs year')
plt.ylabel('no of attacks')
plt.show()
```

# No of Attacks per year

```python
plt.subplots(figsize=(10, 6))
sns.countplot(data=df1, x='iyear', palette='RdYlGn_r', edgecolor=sns.color_palette("YlOrBr", 10))
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
```

# Methods of Attacks

```python
plt.subplots(figsize=(10, 6))
sns.countplot(data=df1, x='attacktype1_txt', palette='RdYlGn_r', edgecolor=sns.color_palette("YlOrBr", 10))
plt.xticks(rotation=90)
plt.title('Methods Of Terrorist Attacks')
```

```python
plt.subplots(figsize=(10, 6))
sns.countplot(data=df1, x='weaptype1_txt', palette='RdYlGn_r', edgecolor=sns.color_palette("YlOrBr", 10))
plt.xticks(rotation=90)
plt.title('weapons Of Terrorist Attacks')
```

```python
plt.subplots(figsize=(10, 6))
sns.barplot(y=df1['gname'].value_counts()[1:12].index,x=df1['gname'].value_counts()[1:12].values,
           palette='copper')
plt.xticks(rotation=90)
plt.title('weapons Of Terrorist Attacks')
```

```python
fig,axes = plt.subplots(figsize=(16,11),nrows=1,ncols=2)
sns.barplot(x = df1['country'].value_counts()[:20].values, y = df1['country'].value_counts()[:20].index, 
            ax=axes[0],palette = 'magma');
axes[0].set_title('Terrorist Attacks per Country')
sns.barplot(x=df1['region_txt'].value_counts().values,y=df1['region_txt'].value_counts().index,
            ax=axes[1])
axes[1].set_title('Terrorist Attacks per Region')
fig.tight_layout()
plt.show()
```
